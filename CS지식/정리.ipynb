## 컴퓨터 구조
- 문제 해결 능력이 달라진다.

단순 문법이 아니라, 컴퓨터의 어떤 로직에서 문제가 발생하는지 찾을 수 있다.
ex) 동일한 코드를 작성하였는데, 오류가 나는경우. 내 컴퓨터에서는 문제는 없지만, 타 환경에서 문제가 생기는 경우.

- 성능,비용,처리에 유용하다.

컴퓨터가 이해하는 정보 : 데이터, 명령어
핵심 부품 : cpu,메모리,보조기억장치,입출력장치
메모리 : 현재 실행하는 데이터,명령어를 저장하는 공간
cpu의 내부 구성
- ALU(계산기)
- 레지스터(CPU내부의 작은 임시저장장치)
- 제어장치(제어신호를 내보내고, 명령어를 해석하는 장치)
    - 컴퓨터의 부품을 관리하고 작동시키는 것(e.g. 메모리 읽기신호, 쓰기 신호)

ex) 메모리 내부 구성
1번지 더하라, 3번지와 4번지를(명령어)
2번지 저장하라, 연산결과를(명령어)
3번지 120(데이터)
4번지 100(데이터)

제어장치(메모리 읽기) - 메모리(1번지의 명령어가 레지스터로 이동) - 제어장치(해석후 데이터를 레지스터에 저장, 이후 ALU로 전송) - ALU에서 더한 뒤 레지스터에 결과값을 저장
- 제어장치(메모리 쓰기 신호를 통해 더한 값인 220을 메모리에 저장)

보조기억장치 : usb,ssd등
메모리는 현재의 명령, 데이터를 저장하지만 전원이 꺼지면 휘발되는 특성이있음.
보조기억장치는 전원이 종료되더라도 기억하기위한 목적

입출력장치 : 모니터, 키보디, 마우스

메인보드 & (시스템) 버스
- 메인보드에 연결된 부품은 버스를 통해 정보를 주고받는다.
- 컴퓨터의 핵심 부품을 연결하는 버스는 시스템 버스

시스템 버스
- 주소버스
- 데이터 버스
- 제어 버스




## 운영체제
시스템 자원
- 프로그램이 실행되기 위해 마땅히 필요한 요소
- 컴퓨터의 네 가지 핵심 부품 포함

운영체제
- 자원을 관리하는 특별한 프로그램
- 프로세스(실행중인 프로그램)을 관리하는 특별한 프로그램
메모리에 저장되어있음.
메모리의 커널영역에 운영체제가 위치하고 있고, 사용자가 이용하는 부분을 사용자영역이라고 하며, 응용프로그램이 실행되는 영역이다.

ex) 메모리
커널 영역
- 운영체제
사용자 영역
- 워드 프로세스 : 2000번지
- 인터넷 브라우저 : 1500번지
- 메모장 : 1000번지

만약 게임을 실행해야한다고 하면, 운영체제가 3000번지에 배정. 또한 메모장을 종료한다면 1000번지에서 종료

만약, 물리적인 메모리공간보다 큰 프로그램을 실행해야할 경우 페이징, 스와핑을 통해 실행
페이징 : 컴퓨터가 메인 메모리에서 사용하기위해 2차 기억장치로부터 데이터를 저장하고 검색하는 메모리 관리 기법

스와핑 : 현재 메모리에서 잠깐 다른 저장공간으로 옮겨졌다가 돌아와서 실행되는것. 
ex) 10개의 프로세스를 실행시킬 수 있는 공간에 이미 10개의 프로세스가 실행중인상태에서 11번째 프로세스를 실행 시켜야한다면, 가장 안쓰는(이벤트를 가장 오래 기다리고있는) 프로세스를 보조 기억장치로 잠시 옮겨두는 것.

운영체제의 자원(cpu) 관리
여러 프로그램들이 cpu를 필요할때, 운영체제가 이를 얼마나 오래, 먼저 실행할지 결정

문지기역할(시스템 호출)을 통한 자원 보호
