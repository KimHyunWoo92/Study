```python
# 문제에서 지역별 범죄률의 전년대비 증감률을 구하라.
# 우리게에 필요한 "발생건수"
# 발생건수만 불리안 인덱싱을 통해서 가져오기
```

```python
dt19 = df2019[df2019["구분"] == "발생건수"]
dt20 = df2020[df2020["구분"] == "발생건수"]
dt21 = df2021_03[df2021_03["구분"] == "발생건수"]
```

```python
dt19
```

```python
# 범죄별로 발생건수가 나눠져 있기떄문에, 이를 통합해줌.
# sum 함수를 사용해서, 모두 더해줄때, 우리는 지역구별 증감률을 보아야하기떄문에 행연산이 필요.
# 데이터 프레임에서 열은 axis = 0 / 행은 axis=1
dt19["총합"] = dt19.sum(axis=1)
```

```python
# 데이터에 "구분" column은 문자열로 되어있기때문에, 문자열과 숫자의 연산이 안됨
# 우리는 범죄의 발생건수의 합만 필요하기때문에 여기서 인덱싱이 필요.
# 컬럼에 대해 인덱싱을 진행
# loc, iloc의 차이 : loc는 ~까지로 인식되어 슬라이싱의 규칙 마지막보다 하나큰 숫자 입력이라는 규칙 예외적용
dt19["총합"] = dt19.loc[:,"살인":"폭력"].sum(axis=1)
dt19
# 해당 오류는 loc를 명시적으로 사용하는걸 권장, 또는 슬라이싱을 할때 
```

```python
dt20.loc[:,"총합"] = dt20.loc[:,"살인":"폭력"].sum(axis=1)
dt21.loc[:,"총합"] = dt21.loc[:,"살인":"폭력"].sum(axis=1)
```

```python
print(dt20.head())
print(dt21.head())
```

```python
df19_end = dt19.loc[:,"총합"]
df20_end = dt20.loc[:,"총합"]
df21_end = dt21.loc[:,"총합"]
```

```python
# series의 name속성을 지정하면, 데이터 프레임으로 합칠때, columns로 자동 할당
# 
df19_end.name = "2019총계"
df20_end.name = "2020총계"
df21_end.name = "2021총계"
```

```python
df19_end
```

```python
# 전년대비 증감률 계산
# 범죄 증감률 = (금년 - 전년)/ 전년 * 100
# 현재 연도가 19,20,21 세가지 이기떄문에 20-19년도 / 21-20 년도에 대한 증감률 계산가능

ch_rate20 = (df20_end-df19_end)/df19_end * 100
ch_rate20.name = "2019-20 증감률"
ch_rate21 = (df21_end - df20_end) / df20_end * 100
ch_rate21.name = "2020-21 증감률"
```

```python
# 만들어진 데이터를 하나로 합치기
# concat함수 사용
# concat 사용시, 중요한 것. 축 설정
# axis = 0
# concat 사용시 대괄호
```

```python
result = pd.concat([df19_end,ch_rate20,df20_end,ch_rate21,df21_end],axis=1)
result
```

```python
# 계산할때 round 함수를 이용하는것이 가장 간편.
# round : 반올림 함수, () 안에 숫자까지 표시, 괄호 안의 숫자+1 자리에서 반올림하여,괄호안의 숫자까지표시
# ch_rate20 = ((df20_end-df19_end)/df19_end * 100)round(2)
```

```python
# 소수점 두번쨰 자리까지 보고싶을떄
result["2019-20 증감률"] = result["2019-20 증감률"].map('{:.2f}'.format)
result
#map 함수는 series의 각 원소에 대해 특정 함수를 적용할 수 있게 해주는 메서드
```

```python
# map은 하나의 series에 접근가능.
# 2개이상의 series에 적용하려면 두번 실행해야함.
# 한번에 2개이상의 열에 접근할떄는 apply 사용하면 좋고, apply의 경우에도
# 각 원소에 대해 소수점을 표시할때는 map을 사용해서 각 원소에 대해 소수점2번째자리까지 변화시키거나.
# def 소수점변환 함수를 통해 적용시켜야함.
```

```python
def format_decimal(x):
    return '{:.2f}'.format(x)
```

```python
result[["2019-20 증감률","2020-21 증감률"]] = result[["2019-20 증감률","2020-21 증감률"]].applymap(format_decimal) 
```
