```python
weight = int(input("현재 몸무게 :"))
target = int(input("목표 몸무게 :"))
week = 1
while True:
    minus_wg = int(input(f"{week}주차 감량몸무게 :"))
    weight -= minus_wg
    if weight <= target:
        print(f"{weight}kg 달성!! 축하합니다!")
        break
    week += 1
```

```python
weight = int(input("현재 몸무게 :"))
target = int(input("목표 몸무게 :"))
week = 1
while weight >= target:
    minus_wg = int(input(f"{week}주차 감량 몸무게 : "))
    weight -= minus_wg
    if weight <= target:
        print(f"{weight}kg 달성!! 축하합니다!")
    week +=1
```

```python
import random
rand_num = random.randint(1,50)
while True:
    num1 = int(input("숫자를 입력하세요 >>"))
    if rand_num > num1:
        print(f"{num1}보다 큰 수 입니다.")
    elif rand_num < num1:
        print(f"{num1}보다 작은 수 입니다.")
    else:
        print("정답을 맞추셨습니다.")
        break
```
