```python
# while 예제 1
# 문제의 단서가 두개의 정수가 0이 들어올 때까지 반복한다고 됨.
```

```python
# 끝나는 조건 : 두개의 정수가 0
# 실행문장이 한번 종료 될때마다 새로 값을 입력하여야함.
while True:
    # 첫번째 정수, 두번째 정수를 입력
    # input() 함수의 안에 문자열은 prompt로 안내메세지를 적을 수 있음
    num1 = int(input("첫 번째 정수 입력 >>"))
    num2 = int(input("두 번째 정수 입력 >>"))
    if num1 == 0 and num2 == 0:
        print("프로그램이 종료 되었습니다.")
        break
    #문제에서 두 정수의 합만 있기 때문에, 0,0이 들어오지 않으면 다 더하면 됨
    #두 정수의 차, 곱 등 여러 조건이 있었다면 elif 를 활용해 조건 추가
    else:
        print(f"두 정수의 합 : {num1+num2}")
```

```python
# break 활용
weight = int(input("현재 몸무게 : "))
target = int(input("목표 몸무게 : "))
# 현재 몸무게와 목표 몸무게가 반복될떄마다 바뀌면 될까?
# 반복될때마다, 다이어트 기간이 증가하는걸 볼 수 있음
week = 1
while True:
    minus_wg = int(input(f"{week}주차 감량 몸무게 : "))
    weight -= minus_wg
    if weight <= target:
        print(f"{weight}kg 달성!! 축하합니다!")
        break    
    week += 1
```

```python
# 조건식으로 접근
# 현재 몸무게가 목표 몸무게 보다 적었을때 끝나야함
# 그럼 현재 몸무게가 목표 몸무게보다 클떄는 계속 다이어트 진행
weight = int(input("현재 몸무게 : "))
target = int(input("목표 몸무게 : "))
week = 1
while weight > target:
    minus_wg = int(input(f"{week}주차 감량 몸무게 : "))
    weight -= minus_wg
    week += 1
print(f"{weight}kg 달성!! 축하합니다!!")
```

```python
# 두개의 로직을 합치면
num = int(input("정수 입력 >>"))
list1 = []
for i in range(1,num+1):
    if num%i == 0:
        list1.append(i)
result = ""
for j in list1:
    result += str(j)+" "
print(f"{num}의 약수 : {result}")
```

```python
# 반복문을 두번이나 사용하고, 고려해야할 것도 많음
# 보기좋게 표시하기 위해, 번거로운과정을 거쳐야함
# 문자열을 연결시켜주는 함수
# join함수 ->  "구분자".join(리스트)
```

```python
num = int(input("정수 입력 >>"))
list1 = []
for i in range(1,num+1):
    if num%i == 0:
        list1.append(str(i))
print(f"{num}의 약수는 {' '.join(list1)}")
```
