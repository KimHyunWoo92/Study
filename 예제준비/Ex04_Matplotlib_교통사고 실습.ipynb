### 기본적인 선 그래프 그려보기

```python
# Line Plot( 선그래프)

y = [2, 12, 6]             # x축은 자동으로 지정됨

# 그래프를 그려주는 역할
# 데이터를 하나의 리스트만 전달하면 x축은 인덱스를 자동부여
plt.plot(y)

# 그동안 그렸던 plot들을 출력하는 역할
# print()와 같은 기능   - 다 그렸다면 show 출력해주기
# 주피터노트북에서는 생략해도 출력되나 다른 프로그램을 위해 씀!

plt.show()
```

```python
# 데이터 2개 넣어보기
# x축, y축 설정해서 그려보기

# 선 스타일을 설정해보자

x = [1,2,3]
y=[2,4,6]
                                                                                                 # marker : 입력한 값에 포인트   '대문자 썼더니 오류남'
plt.plot(x, y, ls= ':' , color='tomato' , marker='x', lw = 5, mec="b", mfc = "y" , ms=15)       # ls = line style     :  , --       color : 라인 색변경
                                                         # lw = line Wieth  marker edge color  / marker face color / marker size 
plt.savefig('line_plot.png')              # 이미지로 저장하고  싶을 때    - 맨아래로 순서를 바꿨더니 그림 안나옴!

plt.show()
```

```python
# 속성값
# 1. 라인의 스타일 설정 : ls(Line Style) -> 문자열
# 2. 라인의 두께 지정 : lw(Line Width) -> 그냥 숫자로 적어주기!! 문자X
# 3. 선의 색 바꾸기 : c(color) -> 선색과 마커의 색까지 모두 변경
# 4. 값의 위치 표시 : marker -> 문자열
# 5. 마커 내부 색깔 : mfc(marker face color) -> 문자열
# 6. 마커 선 색깔 : mec(marker edge color) -> 문자열
# 7. 마커 크기 : ms -> 숫자
# 8. 마커 선 굵기 : mew(marker egde width) -> 숫자
```

```python
import numpy as np
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

#    $ 기호 안에 넣어주고 싶은 기호 넣어주기 가능
plt.plot(x, y , ls="-." , marker="$♣$", lw=4,  mfc='r', mec='g', ms="10")
#marker="$♧$"


# 그래프 숫자 표시 범위 지정(xlim, ylim)
plt.xlim(-2,10)
plt.ylim(-5,12)


# 그리드 그리는 법( 모눈종이, 격자눈금)
plt.grid()


# 눈금 출력방법 -> ticks
plt.yticks(y)



plt.show
```

 ###  교통사고가 가장 많이 발생하는 시간대를 pie 그래프로 시각화 하기!!!

```python
# 시간대 정보 접근하기

# 시간 데이터 뽑아서 time이라는 변수에 담아주기

data["발생년월일시"]                     # int 형 자료형에서 뒤에 두자리만 필요

time =data["발생년월일시"] % 100           #  정수형이므로 연산을 통해 뒤에 두자리만 뽑아냄

time
```

```python
# 시간을 카테고리화 해보자 [ 0~23 시간을 8개의 구간으로 나누자.]

bins=[-1,2,5,8,11,14,17,20,23]   # 초과~ 이하
labels =['0~2','3~5','6~8','9~11','12~14','15~17','18~20','21~23']

pie = pd.cut(time, bins, labels=labels)        # 카테고리화 됨

pie              # 1차원의 Series데이터
```

```python
# 인덱스를 기준으로 정렬

rs_pie = pie.value_counts().sort_index()
rs_pie
```

```python
# pie 차트화
# pie(autopct = '%.2f')    ->  % 문자열을 넣겠다  : 부채꼴 안에 표시될 숫자의 형식을 지정함. 전체 퍼센트 비율을 넣어줌

plt.pie(rs_pie, labels= labels, autopct='%.2f%%')          # % 기호 넣어주고 싶어서 %%

plt.title("시간대별 교통사고 발생건수")
plt.show
```

```python
'''이 코드는 Matplotlib의 plt.pie() 함수를 사용하여 파이 차트를 생성합니다. 
이 함수를 사용하면 데이터를 원형으로 표시하여 각 부분이 전체에 대한 상대적 비율을 나타낼 수 있습니다.

주요 내용은 다음과 같습니다:

rs_pie: 파이 차트에 표시할 데이터입니다. 이는 파이 차트를 생성할 때 각 부분의 크기를 결정합니다.
labels: 파이 차트의 각 부분에 대한 레이블입니다. 이는 데이터의 이름이나 범주를 나타내는데 사용됩니다.
autopct='%.2f%%': 이 매개변수는 각 부분의 비율을 백분율로 표시하는 형식을 지정합니다. 
'%.2f%%'는 소수점 이하 두 자리까지 표시하는 부동 소수점 형식입니다.
따라서 이 코드를 실행하면 rs_pie에 따라 파이 차트가 생성되고, 각 부분에는 레이블이 표시되며 각 부분의 백분율이 소수점 두 자리까지 표시됩니다.'''
```
