```python
ages = [0,2,10,21,23,37,31,61,20,41,32,100]
bins = [0,15,25,35,60,99]
labels = ["미성년자","청년","중년","장년","노년"]
cate = pd.cut(ages,bins,labels=labels)
```

```python
# cut 함수는 카테고리형 배열을 만들어 줄 뿐, 데이터프레임으로 생성해주진 않음
# 따라서, 데이터프레임화 시키려면 추가적인 작업이 필요
cate
```

```python
cate.categories
```

```python
agearr = pd.DataFrame(ages,columns=["ages"])
agearr
```

```python
agearr["구분"] = cate
```

```python
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                    "B":["B0","B1","B2","B3"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]},
                    index=[0,1,2,3])

df2 = pd.DataFrame({"A":["A4","A5","A6","A7"],
                    "B":["B4","B5","B6","B7"],
                    "C":["C4","C5","C6","C7"],
                    "D":["D4","D5","D6","D7"]},
                    index=[4,5,6,7])

df3 = pd.DataFrame({"A":["A8","A9","A10","A11"],
                    "B":["B8","B9","B10","B11"],
                    "C":["C8","C9","C10","C11"],
                    "D":["D8","D9","D10","D11"]},
                    index=[8,9,10,11])
```

```python
result = pd.concat([df1,df2,df3])
result
```

```python
#keys 속성 부여시 다중 index
# 각 데이터에 대한 라벨을 붙여주는 역할
# 여러 데이터를 구분할때 유용함
result = pd.concat([df1,df2,df3],keys=["x","y","z"])
result
```

```python
df4 = pd.DataFrame({"B":["B2","B3","B6","B7"],
                    "D":["D2","D3","D6","D7"],
                    "F":["F2","F3","F6","F7"]},
                    index=[2,3,6,7])
```

```python
result = pd.concat([df1,df4],axis=1)
result
```
