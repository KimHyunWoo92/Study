```python
# 데이터 분석 라이브러리 로드하기(가져오기)

import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기

data2019 = pd.read_csv("2019.csv",encoding='euc-kr' , index_col='관서명')
data2020 = pd.read_csv("2020.csv", encoding='euc-kr' , index_col='관서명')
data2021 = pd.read_csv("2021.csv", encoding='euc-kr' , index_col='관서명')
```

```python
# 특정연도(2021년)에만 존재하는 경찰청인 '광주지방경찰청' 행 삭제   -drop()
```

```python
data2021.drop('광주지방경찰청', inplace = True)      #~할 준비가 되어있는 inplace=True!!!!!!!!
                                                                                    # 혹은 재대입
```

```python
print(data2019.shape)
print(data2020.shape)
print(data2021.shape)
```

```python
# 1. 각  연도별 범죄데이터의 총합을 구해서 새로운 컬럼(열) 생성하기('총합')

data2019['총합'] = data2019.iloc[ : , 1: ].sum(axis=1)     #data2019.loc[ : , '살인' : '폭력'  ].sum(axis=1)
data2019                                                                            #계속하면 총합이 누적되므로 한번만 해야됨
```

```python
data2020['총합']  = data2020.iloc[ : , 1 : ].sum(axis=1)
data2020
```

```python
data2021['총합']  = data2021.iloc[ : , 1 : ].sum(axis=1)
data2021
```

```python
# 2. '구분' 컬럼이 '발생건수' 인 데이터들만 가져오기(불리언 인덱싱)

# temp2019, temp2020, temp2021 변수에 연도별로 각각 담아주기

# 2-1) '구분' 컬럼에 접근하기 - 인덱싱

temp2019 = data2019[data2019.iloc[ : , 0] =='발생건수']

temp2020 =  data2020[data2020[ '구분' ]=='발생건수']

temp2021 =  data2021[data2021[ '구분' ]=='발생건수']
temp2021
```

```python
# 3. 2번에 담아둔 데이터(temp2019)에서 '총합'컬럼을 기준으로 데이터 가져오기

# 데이터 가져오기(Series형태) 
# s2019, s2020, 2021라는 변수에 연도별로 담아주기
s2019 = temp2019['총합']
s2020 = temp2020['총합']
s2021 = temp2021['총합']
s2019
```

```python
# 3-1, 3번에 담아둔 총합 데이터(Series)의 이름을 바꿔주기

#'2019총계'

s2019.name='2019총계'                   # s2019.rename('2019총계') 로 했더니만 병합할 때 이름 반영안됨!!!
s2020.name='2020총계'
s2021.name='2021총계'
```

```python
# 범죄 증감율 계산

# (금년-작년)/작년 *100

# 2019~2020 증감율 구한 뒤 plus2020 변수에 담아주기

# 2020~2021 증감율 구한 뒤 plus2021 변수에 담아주기


plus2020 = (s2020-s2019)/s2019 *100
plus2021 = (s2021-s2020)/s2020 *100

plus2021
```

```python
# 증감율 데이터(Series) 이름 바꿔주기

# plus2020  --> 2019-2020 증감율

plus2020.name ='2019-2020증감율'

# plus2020  --> 2019-2020 증감율

plus2021.name ='2020-2021증감율'

plus2021
```

```python
# 시리즈 병합하기 (concat() 함수 사용)
# concat함수는 시리즈, 데이터프레임을 병합해준다.

pd.concat([s2019, plus2020,s2020, plus2021, s2021], axis=1)
```
