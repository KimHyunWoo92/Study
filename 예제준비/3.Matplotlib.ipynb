```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### 선 그래프

```python
y = [2,4,6]

plt.plot(y) # 그래프 그리기
plt.show() # 그래프 출력
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
# line style
plt.plot(x,y, linestyle = '--')
plt.show()
```

```python
# marker
plt.plot(y, marker = 'o')
plt.show()
```

```python
# line width
plt.plot(x, y , linewidth = 5)
plt.show()
```

```python
# line color
plt.plot(x,y, color = 'r')
plt.show()
```

```python
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15
        , mec = 'g', mew = 5, mfc = 'r')
plt.show()
```

```python
# 범위 지정
# xlim : x 범위 지정
# ylim : y 범위 지정
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15
        , mec = 'g', mew = 5, mfc = 'r')
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.show()
```

```python
# 틱 설정
# xticks : x축 틱 설정 
# yticks : y축 틱 설정
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15
        , mec = 'g', mew = 5, mfc = 'r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

```python
# grid
# 틱의 위치를 잘 보기 위해
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15
        , mec = 'g', mew = 5, mfc = 'r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

```python
# 여러개 그래프 그리기
# 한 그림안에 여러개 그래프 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x) # 그래프 그리기
plt.plot(y)
plt.plot(z)
plt.show() # 위에서 그린 그래프 출력
```

```python
plt.plot(x) # 그래프 그리기
plt.show() # 위에서 그린 그래프 출력
plt.plot(y) 
plt.show() 
plt.plot(z) 
plt.show()
```

```python
# 범례 표시
# label
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv('Traffic_Accident_2017.csv',encoding = 'euc-kr')
data
```

```python
# 요일별 사고 발생건수를 그래프로 그려보자
temp = data['요일'].value_counts()
```

```python
y = temp[['월','화','수','목','금','토','일']]
x = y.index
```

```python
plt.bar(x,y)
plt.ylim(500,620)
plt.show()
```

```python
car_to_car = data[data['사고유형_대분류']=='차대차']
```

```python
car_to_car_by_place = car_to_car[['사상자수','발생지시도']].groupby('발생지시도').sum()
```

```python
car_to_car_by_place
```

```python
x = car_to_car_by_place.index
y = car_to_car_by_place.values
```

```python
plt.figure(figsize=(15,8))
plt.plot(x,y)
plt.show()
```
