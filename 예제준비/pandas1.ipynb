```python
score = pd.DataFrame({
    '과목': ["파이썬", "db", "자바", "크롤링", "web"],
    '1반': [45, 76, 47, 92, 11],
    '2반': [44, 92, 92, 81, 79],
    '3반': [73, 45, 45, 85, 47],
    '4반': [39, 69, 69, 40, 26]
})
```

```python
score.to_csv("score.csv",index=False,encoding="utf-8")
```

```python
score = score.set_index("과목")
```

```python
score
```

```python
# apply() 함수 axis =0이 기본값
# axis =0 열을 기준, axis=1 행 기준으로 반복적으로 적용
```

```python
def max_min(row):
    return row.max()-row.min()
```

```python
result = score.loc[:"web",:"4반"].apply(max_min, axis=1)
print(result)
```

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'A': [9, 10], 'B': [11, 12]})

# 세로로 결합
result = pd.concat([df1, df2, df3], axis=0)
print(result)
```
