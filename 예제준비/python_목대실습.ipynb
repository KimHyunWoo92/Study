```python
# log 데이터 문제
import random

def generate_pw():
    return f"{random.randint(1,9999):04d}"

def generate_log():
    #success
    su_pw = [generate_pw() for _ in range(2)]
    su_front = [int(pw[:2]) for pw in su_pw]
    su_pass = f"{max(su_front):02d}"

    #fail
    fail_pw = [generate_pw() for _ in range(2)]
    fail_back = [int(pw[2:]) for pw in fail_pw]
    fa_pass = f"{min(fail_back):02d}"

    pass_pw = su_pass+fa_pass

    logs = [[su_pw[0], "success"],
            [su_pw[1], "success"],
            [fail_pw[0], "fail"],
            [fail_pw[1], "fail"]]
    
    random.shuffle(logs)
    
    return logs, pass_pw


logs, password = generate_log()
```

```python
print(logs)
```

```python
# log 문제
def solution(logs):
    success_pw = []
    fail_pw = []

    for pw,status in logs:
        if status == "success":
            success_pw.append(pw)
        elif status == "fail":
            fail_pw.append(pw)
    
    su_front = [int(pw[:2]) for pw in success_pw]
    su_pass = f"{max(su_front):02d}"

    fa_back = [int(pw[2:]) for pw in fail_pw]
    fa_pass = f"{min(fa_back):02d}"

    answer = su_pass+fa_pass
        
    return int(answer)
```

```python
print(solution(logs),password)
```

```python
# 운송 문제

import random
import string

def generate_delivery(n=10):
    methods = ["express","normal","slow"]
    items = list(string.ascii_lowercase)

    data = []

    for i in range(n):
        method = random.choice(methods)
        dist = random.randint(1,100)
        item = random.choice(items)

        data.append({
            "method":method,
            "dist":dist,
            "item":item
        })
    return data

datas = generate_delivery()
```

```python
def solution(datas):
    max_cost = {
        "express" : (0,""),
        "normal" : (0,""),
        "slow" : (0,"")
    }
    for data in datas:
        method = data["method"]
        dist = data["dist"]
        item = data["item"]

        if method == "express":
            cost = dist*5 + (dist//10)*10
        elif method == "normal":
            cost = dist*5 + (dist//30)*10
        else:
            cost = dist*1

        if cost > max_cost[method][0]:
            max_cost[method] == (cost,item)
    
    return [max_cost["express"][1],max_cost["normal"][1],max_cost["slow"][1]]
```

```python
# 센서 문제

sensor = {
    "factory": [18, 20, 35, 36, 15, 22, 19, 25, 30, 32, 29, 28],
    "lab":     [19, 22, 40, 38, 25, 20, 21, 23, 30, 34, 35, 33],
    "room":    [21, 23, 30, 29, 28, 28, 29, 30, 30, 31, 32, 32]
}

standard = [
    {"factory": 10},
    {"lab": 15},
    {"room": 8}
]
```

```python
def solution(sensor,standard):
    answer = []
    std = {}
    for tem in standard:
        for key, val in tem.items():
            std[key] = val

    areas = sorted(sensor.keys())

    for area in areas:
        values = sensor[area]
        limit = std[area]
        count = 0

        for i in range(len(values)-1):
            if abs(values[i+1]-values[i]) > limit:
                count +=1
        answer.append(count)
        
    return answer
```

```python
solution(sensor,standard)
```
