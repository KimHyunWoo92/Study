```python
# 도로 종류 및 속도 입력
road = input("도로 종류 입력(고속도로,국도,스쿨존)")
speed = int(input("주행속도 입력(km/h)"))

if road == "고속도로":
    limit = 100
    if speed > limit+20:
        print("찍혔습니다.")
    else:
        print("통과했습니다.")
elif road == "국도":
    limit = 60
    if speed > limit + 10:
        print("찍혔습니다.")
    else:
        print("통과했습니다.")
elif road == "스쿨존":
    limit = 30
    if speed > limit:
        print("찍혔습니다.")
    else:
        print("통과했습니다.")
else:
    print("잘못된 도로를 입력하였습니다.")
    
```

```python
# 커피한잔의 가격은 2500원
# 하루 목표 매출을 입력받고, 손님의 커피주문을 입력받음
# 목표 달성까지 반복하는 예제
# 주문을 할때마다, 매출과 누적매출을 표현
coffee = 2500
goal = int(input("목표 매출 금액 : "))

total = 0

while total <= goal:
    amount = int(input("커피 수량 입력 :"))
    sale = coffee*amount
    total += sale
    print(f"이번 주문 매출 : {sale}, 누적매출 : {total}원")

print(f"목표달성!! 매출은 {total}원 입니다!!")
```

```python
# 100 이하의 완전수 구하기
# 완전수는 자기 자신을 제외한 모든 약수의 합이 자기 자신과 같은 수
list1 = []
for num in range(2,101):
    add = 0
    for i in range(1,num):
        if num%i == 0:
            add += i
    if add == num:
        list1.append(num)
print(list1)
```
